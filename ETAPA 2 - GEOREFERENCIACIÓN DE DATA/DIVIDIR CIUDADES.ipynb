```python
import pandas as pd

OUTPUT_DIR = "ETAPA 2 - GEOREFERENCIACIÓN DE DATA/outputs"

# Leer dataset georreferenciado
df = pd.read_csv(f"{OUTPUT_DIR}/censo_georeferencia.csv", low_memory=False)

# Asegurar string
df["COD_DANE_ANM"] = df["COD_DANE_ANM"].astype(str).str.strip()

# DPTO + MPIO (primeros 5 dígitos)
df["DPTO_MPIO"] = df["COD_DANE_ANM"].str[:5]

# mapa de ciudades
city_map = {
    "63001": "armenia",
    "66001": "pereira",
    "76001": "cali"
}

# generar CSV por ciudad
for code, name in city_map.items():
    
    out = df[df["DPTO_MPIO"] == code].copy()
    
    out_path = f"{OUTPUT_DIR}/censo_georeferencia_{name}.csv"
    
    out.to_csv(out_path, index=False)
    
    print(name, "-> filas:", len(out), "| guardado:", out_path)


# revisar municipios presentes
print("\nMunicipios presentes (DPTO+MPIO):")
print(df["DPTO_MPIO"].value_counts())
```
