```python
import pandas as pd

arm  = pd.read_csv("armenia_base_consolidada_reducida.csv",  low_memory=False)
per  = pd.read_csv("pereira_base_consolidada_reducida.csv",  low_memory=False)
cali = pd.read_csv("cali_base_consolidada_reducida.csv",     low_memory=False)
```

```python
df_all = pd.concat([arm, per, cali], ignore_index=True)
print("Filas totales:", df_all.shape[0])
```

```python
df_all.to_csv(
    "desigualdad_3ciudades_FINAL.csv.gz",
    index=False,
    compression="gzip"
)

print("Archivo final guardado")
```
