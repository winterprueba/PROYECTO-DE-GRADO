```python
import pandas as pd

# Lee tu CSV georreferenciado (el que tiene GEOM_WKT)
df = pd.read_csv("censo_georeferencia.csv", low_memory=False)

# Asegura string
df["COD_DANE_ANM"] = df["COD_DANE_ANM"].astype(str).str.strip()

# DPTO+MPIO (primeros 5 dígitos)
df["DPTO_MPIO"] = df["COD_DANE_ANM"].str[:5]

# Mapa de ciudades (ajusta si tus códigos son distintos)
city_map = {
    "63001": "armenia",
    "66001": "pereira",
    "76001": "cali"
}

# Generar 3 CSV
for code, name in city_map.items():
    out = df[df["DPTO_MPIO"] == code].copy()
    out_path = f"censo_georeferencia_{name}.csv"
    out.to_csv(out_path, index=False)
    print(name, "-> filas:", len(out), "| guardado:", out_path)

# (Opcional) Ver si hay otros municipios en el archivo
print("\nMunicipios presentes (DPTO+MPIO):")
print(df["DPTO_MPIO"].value_counts())
```
